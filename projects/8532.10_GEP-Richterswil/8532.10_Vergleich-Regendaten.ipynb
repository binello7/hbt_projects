## Imports

```python
import pandas as pd
from pathlib import Path
import plotly.graph_objects as go
from hbt_tools import utils_hbt as uh
from plotly.subplots import make_subplots
```

## Settings

```python
path_dir = Path(r"Q:\Projekte\8000-\8500er\8532\8532.10 GEP Richterswil"
    r"\05 Berechnungen Grundlagen\8 Entwässerungskonzept\Grundlagen\Regendaten")

path_wae = path_dir / "Wädenswil-Daten_2013-2022.csv"
path_ric = path_dir / "Niederschlagsdaten_ARA-Richterswil.xlsx"
```

## Read data

```python
wae = uh.read_raindata_meteoschweiz(path_wae, abbr=False)
```

```python
ric = pd.read_excel(path_ric, header=0, index_col=0)
```

## Plots

```python
fig = make_subplots()

fig.add_trace(go.Scatter(x=wae.index, y=wae.iloc[:,0], name="Wädenswil"))
fig.add_trace(go.Scatter(x=ric.index, y=ric.iloc[:,0], name="Richterswil"))

save_path = path_dir / "Plot-Vergleich_wae-ric.html"
fig.write_html(save_path)
```
